#### Project Description

This project is to scrap Hong Kong Hang Seng 50 Index stock data from Yahoo Finance website by downloading them as csv file

# 1 Preliminary

```python
####################
# Kernal: Python 2 #
####################

from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
import pandas as pd
```

# 2 Import Data

## 2.1 import dataset

```python
READ_PATH = "C:/Project/Dynamic Page Manual Crawling_50HSI/Input/"
READ_FILE = READ_PATH + "HSI_50_Stock.csv"
df = pd.read_csv(READ_FILE, encoding= "utf-8")
df
```

## 2.2 data formatting

```python
stock_index = df['Code']
stock_index_str = []
for i in range(len(stock_index)):
    stock_index_str.append(str(stock_index[i]))

for j in range(len(stock_index_str)):
    if (len(stock_index_str[j]) == 1):
        stock_index_str[j] = "000" + stock_index_str[j]
    if (len(stock_index_str[j]) == 2):
        stock_index_str[j] = "00" + stock_index_str[j]
    if (len(stock_index_str[j]) == 3):
        stock_index_str[j] = "0" + stock_index_str[j]

df['Code'] = stock_index_str
df
```

# 3 Web Scraping

## 3.1 set chrome download folder

```python
OUTPUT_PATH = 'C:/Project/Dynamic Page Manual Crawling_50HSI/Output'
chromeOptions = webdriver.ChromeOptions()
prefs = {"download.default_directory" : OUTPUT_PATH}
chromeOptions.add_experimental_option("prefs",prefs)
```

## 3.2 initializing the selenium chrome driver

```python
LOAD_PATH = "C:/Project/Dynamic Page Manual Crawling_50HSI/Chrome Driver/"
LOAD_FILE = LOAD_PATH + 'chromedriver.exe'
browser = webdriver.Chrome(executable_path = LOAD_FILE, chrome_options = chromeOptions)
```

## 3.2 prepare the website links

```python
stock_index_web = df['Code']
web_link = "https://finance.yahoo.com/quote/" + stock_index_web + ".HK/history"
web_link[50] = 'https://finance.yahoo.com/quote/^HSI/history'
web_link
```

## 3.3 define the button xpath

```python
max_button_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/div[1]/span[2]/div/div[1]/span[8]/span'
time_button_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/div[1]/span[2]/span/input'
done_button_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/div[1]/span[2]/div/div[3]/button[1]'
apply_button_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/button'
download_button_xpath = '//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[2]/span[2]/a/span'
```

## 3.4 manual scraping

```python
web_lin = web_link[49:]
for url in web_lin: 
    browser.get(url)
    time.sleep(2)
    browser.implicitly_wait(5)
    browser.find_element_by_xpath(time_button_xpath).click()
    browser.find_element_by_xpath(max_button_xpath).click()
    browser.find_element_by_xpath(done_button_xpath).click()
    browser.find_element_by_xpath(apply_button_xpath).click()
    time.sleep(2)
    browser.implicitly_wait(5)
    browser.find_element_by_xpath(download_button_xpath).click()
    time.sleep(2)
```

## 3.5 exit from the browser

```python
browser.quit()
```
